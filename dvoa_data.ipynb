```python
import pandas as pd
from IPython.display import display, HTML
```

```python
wk = '4'
off_path = r'C:\Users\nickp\OneDrive\Desktop\Home Data Projects\FF\dvoa_data_work\team_offense_dvoa.csv'
def_path = r'C:\Users\nickp\OneDrive\Desktop\Home Data Projects\FF\dvoa_data_work\team_defense_dvoa.csv'
schedule_path = r'C:\Users\nickp\OneDrive\Desktop\Home Data Projects\FF\dvoa_data_work\schedule_2023.csv'

off_dvoa = pd.read_csv(off_path)
def_dvoa = pd.read_csv(def_path)
sched = pd.read_csv(schedule_path)

off_dvoa = off_dvoa[['TEAM','DVOA', 'PASS','RUSH']]
def_dvoa = def_dvoa[['TEAM','DVOA', 'PASS','RUSH']]
def_dvoa = def_dvoa.add_prefix('def_')
schedWk = sched[['Team',wk]]
schedWk.rename(columns={schedWk.columns[1]: "OPP" }, inplace = True)

df = pd.merge(off_dvoa, schedWk, left_on='TEAM', right_on='Team', how='left').drop('Team', axis=1)
df = pd.merge(df, def_dvoa, left_on='OPP', right_on='def_TEAM', how='left').drop('def_TEAM', axis=1)

df['ovrl_DVOA'] = df['DVOA'] + df['def_DVOA']
df['ovrl_PASS'] = df['PASS'] + df['def_PASS']
df['ovrl_RUSH'] = df['RUSH'] + df['def_RUSH']

overall = df.sort_values(by='ovrl_DVOA', ascending=False).head(10)
passing = df.sort_values(by='ovrl_PASS', ascending=False).head(10)
rushing = df.sort_values(by='ovrl_RUSH', ascending=False).head(10)
defense = df.sort_values(by='ovrl_DVOA', ascending=True).head(10)

display(HTML(overall.to_html()))
display(HTML(passing.to_html()))
display(HTML(rushing.to_html()))
display(HTML(defense.to_html()))
display(HTML(df.to_html()))
```
